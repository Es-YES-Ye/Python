```python
#y축에 표현할 값을 지정 + 디폴트 색상 확인
ys_str=[]
for i in range(1,31):
  plt.plot([i, i])
  if i%3==0:
    ys_str.append(str(i)) #0,3,6,9 축값을 넣고
  else:
    ys_str.append("  ") #나머지는 0으로 
plt.title("default color, set yticks") 
plt.yticks(range(1,31), ys_str) #10개의 데이터까지는 자동으로 색상이 다르게 부여된다. 
plt.show()
```

```python
#에러 값을  표현하기, 유효한 데이터를 확인할 수 있다. 수직그래프
colors = ['red', 'green', 'blue'] 
names = ['a', 'b', 'c']
income=[100, 160, 130]
err = [2,3,10]
plt.bar(names,income, color=colors, yerr=err)
plt.title("error")
plt.show()
```

```python
#에러 값을  표현하기, 유효한 데이터를 확인할 수 있다. 수평그래프
colors = ['red', 'green', 'blue'] 
names = ['a', 'b', 'c']
income=[100, 160, 130]
err = [2,3,10]
plt.barh(names,income, color=colors, xerr=err)
plt.title("error")
plt.show()
```

```python
#에러 값도 색상 지정
colors = ['red', 'green', 'blue'] 
ecolors = ['g', 'b', 'r']
names = ['a', 'b', 'c']
income=[100, 160, 130]
err = [2,3,10]
plt.bar(names,income, color=colors, yerr=err, ecolor=ecolors)
plt.title("error")
plt.show()
```

```python
#에러 값을 I / H 형태로 표시, 캡 달기
#캡의 색상은 하나로 통일
colors = ['red', 'green', 'yellow'] 
names = ['a', 'b', 'c']
income=[100, 160, 130]
err = [2,3,10]
plt.bar(names,income, color=colors, yerr=err, ecolor='black', capsize=4.0)
plt.title("error")
plt.show()
```

```python
#에러 값을 I / H 형태로 표시, 캡 달기
#캡의 색상은 하나로 통일
colors = ['red', 'green', 'yellow'] 
names = ['a', 'b', 'c']
votes=[200, 260, 230]
err = [2,3,10]
plt.bar(names,votes, color=colors, yerr=err, ecolor='black', capsize=4.0)
plt.title("error")
plt.show()
```

```python
#퍼센트 표시하기
locales = {'a', 'b', 'c', 'd', 'e'}
incomes = [2300, 1200, 1800, 800, 400]
exs=[0, 0.1, 0, 0, 0]
plt.pie(incomes,labels=locales,explode=exs, autopct="%.2f")
plt.title("incomes")
plt.show()
```

```python
#퍼센트 표시할 위치 지정하기
locales = {'a', 'b', 'c', 'd', 'e'}
incomes = [2300, 1200, 1800, 800, 400]
exs=[0, 0.1, 0, 0, 0]
plt.pie(incomes,labels=locales,explode=exs, autopct="%.2f", pctdistance=0.8)
plt.title("incomes")
plt.show()
```

```python
#그림자 효과
locales = {'a', 'b', 'c', 'd', 'e'}
incomes = [2300, 1200, 1800, 800, 400]
exs=[0, 0.1, 0, 0, 0]
plt.pie(incomes,labels=locales,explode=exs, autopct="%.2f", pctdistance=0.8, shadow=True)
plt.title("incomes")
plt.show()
```

```python
#label 위치 지정
locales = {'a', 'b', 'c', 'd', 'e'}
incomes = [2300, 1200, 1800, 800, 400]
exs=[0, 0.1, 0, 0, 0]
plt.pie(incomes,labels=locales,explode=exs, autopct="%.2f", pctdistance=0.8, labeldistance=1.2)
plt.title("incomes")
plt.show()
```

```python
#시계 방향으로 각도 계산하기 
#현재 3시가 기준 각도다. 거기서 90도 지점으로 startangle을 넣으면 12시 기준으로 시작된다.
locales = {'a', 'b', 'c', 'd', 'e'}
incomes = [2300, 1200, 1800, 800, 400]
exs=[0, 0.1, 0, 0, 0]
plt.pie(incomes,labels=locales,explode=exs, autopct="%.2f", startangle=90, counterclock=False) 
#default는 역시계방향이고 false로 지정하면 시계방향으로 돌아간다
plt.title("incomes")
plt.show()
```
